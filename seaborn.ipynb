# Seaborn Data Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

**Pairplot()** which is a function commonly associated with data visualization libraries in Python, such as Seaborn or Matplotlib. Specifically, pairplot() is a function in Seaborn that creates a matrix of scatterplots for data visualization. It allows you to quickly visualize relationships between variables in a dataset.

```python
sns.pairplot(iris, hue='species', height=2.4)
```

## 2.View categories

```python
titanic = sns.load_dataset('titanic')
titanic.drop(['alone', 'alive', 'who', 'adult_male', 'embark_town', 'class'], axis=1, inplace=True)
titanic.dropna(axis=0, inplace=True)
titanic.head()
```

**catplot()** :is a function in the Seaborn library used for creating categorical plots. It can be used to visualize the relationship between one categorical variable and one numeric variable, or to show the distribution of a single categorical variable.

```python
sns.catplot(x='survived', y='age', data=titanic, hue='sex', height=4.5)
```

**boxplot()** : A boxplot is a statistical visualization that provides a summary of the distribution of a numeric variable for different categories. In Seaborn, you can create a boxplot using the **sns.boxplot()** function. 

```python
plt.figure(figsize=(32, 8))
sns.boxplot(x='age', y='fare', data=titanic, hue='sex')
```

## 3.Visualizing distributions

**sns.displot** : function in Seaborn is a figure-level function that allows you to create various types of distribution plots, including histograms, kernel density estimates (KDE), and combinations of these.

```python
sns.displot(titanic['fare'],kde=True, height=3, aspect=2)
```

**sns.histplot()** :  function in Seaborn is a versatile and feature-rich method for creating histograms with added flexibility for customization. Here are a few reasons why you might choose to use histplot():

Integration with Seaborn Aesthetics: histplot() is designed to work seamlessly with Seaborn's aesthetics, allowing you to easily incorporate color palettes, plot styling, and other features available in Seaborn.

Enhanced Customization: Seaborn provides additional parameters that allow for enhanced customization of the histogram. For example, you can easily add a kernel density estimate (KDE) using the kde parameter or adjust the bin width using binwidth.

```python
plt.figure(figsize=(6,4))
sns.histplot(titanic['fare'], kde=True)
plt.show()
```

**jointplot()** is a versatile function in Seaborn that allows you to visualize the relationship between two variables. It combines scatter plots for numeric variables and histograms for each variable along the axes. This makes it useful for exploring the joint distribution of variables and understanding their individual distributions.

```python
sns.jointplot(x='age', y='fare', data=titanic, kind='hex',height=4.8, ratio=3)
plt.show()
```

**sns.heatmap()** : is a Seaborn function that is used to plot rectangular data as a color-encoded matrix. It is particularly useful for visualizing the correlation matrix of a DataFrame or any other 2D dataset.

```python
numeric_titanic = titanic.select_dtypes(include=['float64', 'int64'])

# Creating the heatmap of the correlation matrix
sns.heatmap(numeric_titanic.corr(), annot=True, cmap='coolwarm')
plt.show()
```
